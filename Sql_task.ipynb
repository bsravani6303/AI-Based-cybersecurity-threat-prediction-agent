# **SQL Task**

Prompt:Create tables for a simple college database
Insert some sample data into all tables


```python
import sqlite3
import pandas as pd


conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER,
    grade TEXT
)
""")
cursor.execute("""
CREATE TABLE courses (
    course_id INTEGER PRIMARY KEY AUTOINCREMENT,
    course_name TEXT,
    student_id INTEGER,
    FOREIGN KEY(student_id) REFERENCES students(id)
)
""")

conn.commit()
print("✅ Tables created successfully!")
```

**Prompt:Insert some sample data**

```python
students_data = [
    ('Alice', 20, 'A'),
    ('Bob', 22, 'B'),
    ('Charlie', 21, 'A'),
    ('David', 23, 'C'),
    ('Eva', 22, 'B')
]

courses_data = [
    ('Math', 1),
    ('Science', 1),
    ('History', 2),
    ('Math', 3),
    ('Art', 4),
    ('Science', 5)
]

cursor.executemany("INSERT INTO students (name, age, grade) VALUES (?, ?, ?)", students_data)
cursor.executemany("INSERT INTO courses (course_name, student_id) VALUES (?, ?)", courses_data)
conn.commit()
print("✅ Sample data inserted!")
```

Promt:Display all students Display all courses

```python
pd.read_sql_query("SELECT * FROM students", conn)
pd.read_sql_query("SELECT* FROM students",conn)
```

 Prompt: Retrieve all students who got grade 'A'

```python

pd.read_sql_query("SELECT name, grade FROM students WHERE grade = 'A'", conn)
```

Prompt: Show each student with their enrolled course(s)

```python

pd.read_sql_query("""
SELECT students.name, courses.course_name
FROM students
JOIN courses ON students.id = courses.student_id
""", conn)
```

Prompt: Count how many students are in each grade

```python
pd.read_sql_query("""
SELECT grade, COUNT(*) AS total_students
FROM students
GROUP BY grade
""", conn)
```

 Prompt: Update Bob’s grade to 'A'

```python

cursor.execute("UPDATE students SET grade = 'A' WHERE name = 'Bob'")
conn.commit()
pd.read_sql_query("SELECT * FROM students WHERE name = 'Bob'", conn)
```

 Prompt: Delete students who got grade 'C'

```python
cursor.execute("DELETE FROM students WHERE grade = 'C'")
conn.commit()
pd.read_sql_query("SELECT * FROM students", conn)
```
